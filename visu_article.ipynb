```python
from tools.tools_preprocessing import (
    load_image,
    canny_detector
)
import cv2
import matplotlib.pyplot as plt
from rembg import remove
```

```python
path_A = 'dataset/asl_alphabet_train/A/8.jpg'
path_2 = 'dataset/large_dataset/test_raw/A22_jpg.rf.f02ad8558ce1c88213b4f83c0bc66bc8.jpg'

img_rgb = load_image(path_2, show=False)
# img_rgb = open(path_A)

# img=cv2.imread(path_A,1)
# img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rgb_no_background = remove(img_rgb)
rgb_contour = canny_detector(img_rgb)
rgb_contour_no_back = canny_detector(rgb_no_background)

img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
hsv_no_background = remove(img_hsv)
hsv_contour = canny_detector(img_hsv)
hsv_contour_no_back = canny_detector(hsv_no_background)

img_xyz = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2XYZ)
xyz_no_background = remove(img_xyz)
xyz_contour = canny_detector(img_xyz)
xyz_contour_no_back = canny_detector(xyz_no_background)

img_lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)
lab_no_background = remove(img_lab)
lab_contour = canny_detector(img_lab)
lab_contour_no_back = canny_detector(lab_no_background)

```

```python
fig, axs = plt.subplots(ncols=4, layout='constrained')
for ax in axs.ravel():
    ax.set_axis_off()

axs[0].imshow(img_hsv)
axs[0].set_title("Original image", fontsize=10)

axs[1].imshow(hsv_no_background)
axs[1].set_title("Removed background", fontsize=10)

axs[2].imshow(hsv_contour)
axs[2].set_title("Outlined image", fontsize=10)

axs[3].imshow(hsv_contour_no_back)
axs[3].set_title("Outlined no \n background image", fontsize=10)

fig.suptitle('HSV mode', fontsize=16, y=0.75)

fig.savefig('HSV_mode_2.png')
```

```python
fig, axs = plt.subplots(ncols=4, layout='constrained')
for ax in axs.ravel():
    ax.set_axis_off()

axs[0].imshow(img_rgb)
axs[0].set_title("Original image", fontsize=10)

axs[1].imshow(rgb_no_background)
axs[1].set_title("Removed background", fontsize=10)

axs[2].imshow(rgb_contour)
axs[2].set_title("Outlined image", fontsize=10)

axs[3].imshow(rgb_contour_no_back)
axs[3].set_title("Outlined no \n background image", fontsize=10)

fig.suptitle('RGB mode', fontsize=16, y=0.75)

fig.savefig('RGB_mode_2.png')
```

```python
fig, axs = plt.subplots(ncols=4, layout='constrained')
for ax in axs.ravel():
    ax.set_axis_off()

axs[0].imshow(img_xyz)
axs[0].set_title("Original image", fontsize=10)

axs[1].imshow(xyz_no_background)
axs[1].set_title("Removed background", fontsize=10)

axs[2].imshow(xyz_contour)
axs[2].set_title("Outlined image", fontsize=10)

axs[3].imshow(xyz_contour_no_back)
axs[3].set_title("Outlined no \n background image", fontsize=10)

fig.suptitle('XYZ mode', fontsize=16, y=0.75)

fig.savefig('XYZ_mode_2.png')
```
