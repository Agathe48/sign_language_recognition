```python
from tools.tools_preprocessing import *
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loadImage(src, show = True):
    img=cv2.imread(src,1)
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    if show : 
        plt.imshow(rgb,interpolation='nearest')
        plt.show()
    return rgb
```

```python
path_A = 'dataset/asl_alphabet_train/A/1.jpg'

img = loadImage(path_A)
```

```python
img_gray = gray_scale(img)
img_canny = canny_detector(img_gray, 20, 100)
plot_canny_img(img, img_canny)
```

```python
img = loadImage(path_A, show = False)
img_gray = gray_scale(img)
img_laplacian = laplacian_detector(img_gray)
sobelx = sobelx_detector(img_gray)
sobely = sobely_detector(img_gray)

plt.subplot(2,2,1),plt.imshow(img_gray,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(img_laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```
